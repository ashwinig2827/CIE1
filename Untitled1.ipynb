```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Iris.csv")
plt.figure(figsize=(8,8))
plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'])
plt.tittl=("Petal Length vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Iris.csv', names=['SepalLengthCm', 'SepalwidthCm', 'PetalLengthCm', 'PetalwidthCm'])
plt.figure(figsize=(8,6))
plt.hist(df['PetalLengthCm'], bins=5, color='red', edgecolor='black')
plt.title('Distribution of Petal Length',fontsize=30)
plt.xlabel('Petal Length (cm)',fontsize=13)
plt.ylabel('Frequency', fontsize=13)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
